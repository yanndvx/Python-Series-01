## Ecg Prototype data processing pipeline 

```python
import numpy as np 
import matplotlib.pyplot as plt

#look at the data
data = np.loadtxt('ECGu.txt')
print (data)
print (data.shape)


#Convertion to mV : ECGmV = ECGraw * 1024V
ECGmV = data*1024
print (ECGmV)

fs = 1000  # Example: 1000 samples per second
t = np.arange(len(data)) / fs
plt.plot(t, ECGmV)
plt.xlabel("Time (s)")
plt.ylabel("ECG (mV)")
plt.show()
```
